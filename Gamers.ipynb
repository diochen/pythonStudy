```python
import requests, time
from bs4 import BeautifulSoup


res = requests.get('https://forum.gamer.com.tw/B.php?bsn=26742')
soup = BeautifulSoup(res.text, 'lxml')

baseURI = "https://forum.gamer.com.tw/"
items = []
for item in soup.select('.b-list .b-list__row'):
    
    topicTitle = item.select_one('.b-list__main .b-list__main__title').text
    topicLink = item.select_one('.b-list__main a').get('href')
    print(topicTitle)
    print(topicLink)
    print("=======================================================")

    for link in item.select('.b-list__main .b-list__main__pages a'):
        print(link.get('href'))
        print(link.text)
    
    print("=======================================================")
    pageRes = requests.get(baseURI + topicLink)
    pageSoup = BeautifulSoup(pageRes.text, 'lxml')
    content = pageSoup.select_one('.c-article__content')
    print(content.text)
    print("=======================================================")
    
    time.sleep(3)
    #break
    
    #desc     = item.select_one('.des').text
    #print(''.join(item.select_one('.price').text.split()) )
    #price    = item.select_one('.price').text.split()[1] 
    #items.append({'headings':headings, 'desc':desc, 'price': price})
    #print(headings, desc, price)
```
